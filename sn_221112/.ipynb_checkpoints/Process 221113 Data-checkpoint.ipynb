```python
import numpy as np
import matplotlib.pyplot as plt
import kpicdrp.data as drp_data
from glob import glob
import os
from astropy.io import fits

plt.rcParams.update({'font.size': 16})
```

```python
#get on axis flux for normalization
onaxis_dir = "onaxis/fluxes"
oa_filelist = glob(os.path.join(onaxis_dir, "*.fits"))
oa_specs = drp_data.Dataset(filelist=oa_filelist, dtype=drp_data.Spectrum)
oa_trace_flux_all = np.array(oa_specs.get_dataset_attributes("fluxes"))
oa_trace_errs_all = np.array(oa_specs.get_dataset_attributes("errs"))
oa_sf4_flux = oa_trace_flux_all[:,0,5,:]
oa_sf4_flux = np.nanmean(oa_sf4_flux,axis=0)
oa_flux_sum = np.nansum(oa_sf4_flux)
oa_sf4_errs = oa_trace_errs_all[:,0,5]
oa_sf4_errs = np.nanmean(oa_sf4_errs,axis=0)
oa_tint = 29.5056

print(oa_flux_sum)
plt.figure()
plt.semilogy(oa_sf4_flux)
plt.semilogy(oa_sf4_errs)
```

```python
data_dir = "jxuan_fluxes/sf4_selected_cycles"
filelist = glob(os.path.join(data_dir, "*.fits"))

planet_specs = drp_data.Dataset(filelist=filelist, dtype=drp_data.Spectrum)
# These have shape (nframes, ntrace, norders, 2048)
all_trace_flux = np.array(planet_specs.get_dataset_attributes("fluxes"))
all_trace_e_flux = np.array(planet_specs.get_dataset_attributes("errs"))

seq_tint = 59.0112
```

```python
wvs = drp_data.Wavecal(filepath="wavecal/20221113_HIP110882_psg_wvs.fits")
wvsol = np.array(wvs.wvs)
frame_nums = np.array(planet_specs.get_header_values(key='FRAMENUM'))

sf4_flux = all_trace_flux[:,3,5,:]
sf4_errs = all_trace_e_flux[:,3,5,:]

sf4_flux_sum = np.nansum(sf4_flux,axis=1)
sf4_errs_sum = np.sqrt(np.nansum(np.square(sf4_errs),axis=1))

sf4_rc = sf4_flux_sum/seq_tint/(oa_flux_sum/oa_tint)
sf4_rc_errs = sf4_errs_sum/seq_tint/(oa_flux_sum/oa_tint)

#sort by frame num
f_idcs = np.argsort(frame_nums)

sf4_rc = sf4_rc[f_idcs]
sf4_rc_errs = sf4_rc_errs[f_idcs]

np.save("sf4_rcs.npy",sf4_rc)
np.save("sf4_rc_errs.npy",sf4_rc_errs)

test_spectrum = np.ravel(sf4_flux[0])
test_errs = np.ravel(sf4_errs[0])
test_wavels = np.ravel(wvsol[0,5])

#sort
sf4_flux = sf4_flux[f_idcs]
sf4_errs = sf4_errs[f_idcs]
```

```python
#bin order 5 data
#num bins must be a factor of 2048
def bin_data_prop_errs(flux,errs,wavels,num_bins):
    bin_size = int(len(flux)/num_bins)
    flux_binned = np.zeros(num_bins)
    errs_binned = np.zeros(num_bins)
    wavels_binned = np.zeros(num_bins)
    for n in range(num_bins):
        flux_subset = flux[n*bin_size:(n+1)*bin_size]
        errs_subset = errs[n*bin_size:(n+1)*bin_size]
        wavels_subset = wavels[n*bin_size:(n+1)*bin_size]
        flux_binned[n] = np.nansum(flux_subset)
        errs_binned[n] = np.sqrt(np.nansum(np.square(errs_subset)))
        wavels_binned[n] = np.nanmean(wavels_subset)
    return flux_binned, errs_binned, wavels_binned
```

```python
oa_flux_binned, oa_errs_binned, oa_wavels_binned = bin_data_prop_errs(oa_sf4_flux,oa_sf4_errs,test_wavels,num_bins=32)

rcs_list = []
errs_list = []

for n in range(26):
    test_spectrum = sf4_flux[n]
    test_errs = sf4_errs[n]
    flux_binned, errs_binned,_ = bin_data_prop_errs(test_spectrum,test_errs,test_wavels,num_bins=32)
    sf4_rc_spec = flux_binned/seq_tint/(oa_flux_binned/oa_tint)
    sf4_rc_spec_errs = errs_binned/seq_tint/(oa_flux_binned/oa_tint)
    rcs_list.append(sf4_rc_spec)
    errs_list.append(sf4_rc_spec_errs)
    
rcs_list = np.array(rcs_list)
errs_list = np.array(errs_list)

print(np.mean(rcs_list[0]))

plt.figure()
plt.semilogy(rcs_list[0])
#plt.semilogy(errs_list[0])
plt.semilogy(rcs_list[-1])

print(oa_wavels_binned.shape)

np.save('o5_rcs_binned.npy',rcs_list)
np.save('o5_errs_binned.npy',errs_list)
np.save('o5_wavels_binned.npy',oa_wavels_binned)
```

```python
#construct a_dm and phi_dm series
cycle1 = fits.getheader('dm_info/UT04_32_46_dmmap0_sf4.fits')
beta = float(cycle1['BETA'])
gain = 0.25
probeamp1 = float(cycle1['PROBEAMP'])*beta
phinull1 = float(cycle1['PHI_S'])+np.pi
nullamp1 = float(cycle1['NULLAMP'])*gain*beta

cycle2 = fits.getheader('dm_info/UT04_32_46_dmmap1_sf4.fits')
probeamp2 = float(cycle2['PROBEAMP'])*beta
phinull2 = float(cycle2['PHI_S'])+np.pi
nullamp2 = float(cycle2['NULLAMP'])*gain*beta

cycle3 = fits.getheader('dm_info/UT04_32_46_dmmap2_sf4.fits')
probeamp3 = float(cycle3['PROBEAMP'])*beta
phinull3 = float(cycle3['PHI_S'])+np.pi
nullamp3 = float(cycle3['NULLAMP'])*gain*beta

cycle4 = fits.getheader('dm_info/UT04_32_46_dmmap3_sf4.fits')
probeamp4 = float(cycle4['PROBEAMP'])*beta
phinull4 = float(cycle4['PHI_S'])+np.pi
nullamp4 = float(cycle4['NULLAMP'])*gain*beta

cycle5 = fits.getheader('dm_info/UT04_32_46_dmmap4_sf4.fits')
probeamp5 = float(cycle5['PROBEAMP'])*beta
phinull5 = float(cycle5['PHI_S'])+np.pi
nullamp5 = float(cycle5['NULLAMP'])*gain*beta
```

```python
nframes = 26
phase_cycle = np.array([0,np.pi/2,np.pi,3*np.pi/2])

a_dm = np.empty(nframes)
phi_dm = np.zeros(nframes)

#cycle 1
a_dm[0] = 0
a_dm[1:5] = probeamp1
a_dm[5] = 0
phi_dm[1:5] = phase_cycle

#cycle 2
efield1 = nullamp1*np.exp(1j*phinull1)
a_dm[6:10] = np.abs(efield1+probeamp2*np.exp(1j*phase_cycle))
phi_dm[6:10] = np.angle(efield1+probeamp2*np.exp(1j*phase_cycle))
a_dm[10] = np.abs(efield1)
phi_dm[10] = np.angle(efield1)

#cycle 3
efield2 = efield1 + nullamp2*np.exp(1j*phinull2)
a_dm[11:15] = np.abs(efield2+probeamp3*np.exp(1j*phase_cycle))
phi_dm[11:15] = np.angle(efield2+probeamp3*np.exp(1j*phase_cycle))
a_dm[15] = np.abs(efield2)
phi_dm[15] = np.angle(efield2)

#cycle 4
efield3 = efield2 + nullamp3*np.exp(1j*phinull3)
a_dm[16:20] = np.abs(efield3+probeamp4*np.exp(1j*phase_cycle))
phi_dm[16:20] = np.angle(efield3+probeamp4*np.exp(1j*phase_cycle))
a_dm[20] = np.abs(efield3)
phi_dm[20] = np.angle(efield3)

#cycle 5
efield4 = efield3 + nullamp4*np.exp(1j*phinull4)
a_dm[21:25] = np.abs(efield4+probeamp5*np.exp(1j*phase_cycle))
phi_dm[21:25] = np.angle(efield4+probeamp5*np.exp(1j*phase_cycle))
a_dm[25] = np.abs(efield4)
phi_dm[25] = np.angle(efield4)

plt.figure()
plt.subplot(2,1,1)
plt.plot(range(26),a_dm,color='k')
plt.xlabel('Frame')
plt.ylabel(r'$a_{DM}$')
plt.title('DM History')

plt.subplot(2,1,2)
plt.plot(range(26),phi_dm,color='k')
plt.xlabel('Frame')
plt.ylabel(r'$\phi_{DM}$')

plt.tight_layout()
plt.savefig('../post_processing/figures/221113_dm_history.png')

#np.save('a_dm_selected.npy',a_dm)
#np.save('phi_dm_selected.npy',phi_dm)
```
