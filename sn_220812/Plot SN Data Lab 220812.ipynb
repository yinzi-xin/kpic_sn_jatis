```python
import numpy as np
from astropy.io import fits
import matplotlib.pyplot as plt
import glob
from scipy import signal
plt.rc('xtick', labelsize=14)
plt.rc('ytick', labelsize=14) 
```

```python
dir_1 = 'null_logs_220812/run_1_10000/'
glob1 = sorted(glob.glob(dir_1+'speckle_params*.fits'))
run_1 = []
for filename in glob1:
    hdul = fits.open(filename)
    run_1.append(hdul[0].data)
run_1 = np.array(run_1)

dir_2 = 'null_logs_220812/run_2_20000/'
glob2 = sorted(glob.glob(dir_2+'speckle_params*.fits'))
run_2 = []
for filename in glob2:
    hdul = fits.open(filename)
    run_2.append(hdul[0].data)
run_2 = np.array(run_2)

dir_3 = 'null_logs_220812/run_3_60000/'
glob3 = sorted(glob.glob(dir_3+'speckle_params*.fits'))
run_3 = []
for filename in glob3:
    hdul = fits.open(filename)
    run_3.append(hdul[0].data)
run_3 = np.array(run_3)
```

```python
plt.figure(dpi=400)
plt.semilogy(np.arange(0,6),run_1[:,1],linewidth=0,marker='o',label='tint=10s',color='k',alpha=0.5)
plt.semilogy(np.arange(6,8),run_2[1:,1],linewidth=0,marker='^',label='tint=20s',color='lightcoral',alpha=0.75)
plt.semilogy(np.arange(8,11),run_3[:,1],linewidth=0,marker='s',label='tint=60s',color='dodgerblue',alpha=0.75)
plt.ylim([1e-5,1e-2])
plt.xlabel('Iteration',fontsize=16)
plt.ylabel('Raw Contrast',fontsize=16)
plt.grid(True)
plt.legend()
plt.title('Off-sky Lab Test (Aug 12, 2022)',fontsize=16)
plt.tight_layout()
```
